```python
df_gram_neg = pd.read_csv('gram_neg.csv')
df_gram_neg = df_gram_neg.astype(object)
df_gram_pos = pd.read_csv('gram_pos.csv')
df_gram_pos = df_gram_pos.astype(object)
```

```python
# Identify common and unique columns
common_columns = df_gram_neg.columns.intersection(df_gram_pos.columns).to_list()
unique_columns1 = df_gram_neg.columns.difference(df_gram_pos.columns).to_list()
unique_columns2 = df_gram_pos.columns.difference(df_gram_neg.columns).to_list()
```

```python
# Merge DF on the common key
merged_df = pd.merge(df_gram_neg, df_gram_pos, on=common_columns, how='outer', suffixes=('_neg', '_pos'))
```

```python
# Include unique columns
for col in unique_columns1:
    merged_df[col] = df_gram_neg.set_index('Organism')[col].reindex(merged_df['Organism']).values

for col in unique_columns2:
    merged_df[col] = df_gram_pos.set_index('Organism')[col].reindex(merged_df['Organism']).values
```

```python
merged_df = merged_df.fillna('-')
merged_df.info()
```
